# Ceasar Cipher
A cipher is used for encrypting messeages for millitary uses. The most basic form of cipher is a ceasar cipher. A ceaser cipher is a cipher where all letters are shifted down by x amount. For example if my message is **Hi**, and the x (also known as key) is 1, then H in hi would be shifted down by 1, making it I. Then the i, would be shifted down by 1, making it j. The ceasar cipher was invented before the mideval ages, by Julius Cesar.

# Subsitution Cipher
During the mideval ages however, ciphers were mainly used for fun, like puzzles. People in the mideval ages mainly used a subsitution cipher. A subsitution cipher, like its main suggests, subsitutes one letter for another. For example, in the message Hi, H could be subsituted for q and i can be subsituted for w, making the encrypted message qi.
